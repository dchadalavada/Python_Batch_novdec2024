# Arithmetic Operators, Exponent Notation, and Compound Operators in Python

## 1. Arithmetic Operators in Python
- Arithmetic operators are used to perform mathematical operations on numeric values.
- Operators: 
  - `+` (Addition)
  - `-` (Subtraction)
  - `*` (Multiplication)
  - `/` (Division)
  - `%` (Modulus)
  - `**` (Exponentiation)
  - `//` (Floor Division)

## 2. Exponent Notation in Python
- Exponent notation is a shorthand for representing large numbers or scientific notation.
- In Python, `e` can be used in numeric literals to represent powers of 10 (e.g., `1e3` represents `1 * 10^3 = 1000`).

## 3. Compound Assignment Operators
- These operators combine arithmetic operations with assignment.
- Examples include:
  - `+=` (Add and assign)
  - `-=` (Subtract and assign)
  - `*=` (Multiply and assign)
  - `/=` (Divide and assign)
  - `**=` (Exponentiate and assign)
  - `//=` (Floor divide and assign)


```python
# 1. Arithmetic Operators in Python

# Example of Addition (+)
x = 10
y = 5
print(f"x + y = {x + y}")  # Output: 15

# Example of Subtraction (-)
print(f"x - y = {x - y}")  # Output: 5

# Example of Multiplication (*)
print(f"x * y = {x * y}")  # Output: 50

# Example of Division (/)
print(f"x / y = {x / y}")  # Output: 2.0

# Example of Modulus (%)
print(f"x % y = {x % y}")  # Output: 0 (since 10 is perfectly divisible by 5)

# Example of Exponentiation (**)
print(f"x ** y = {x ** y}")  # Output: 100000 (10^5)

# Example of Floor Division (//)
print(f"x // y = {x // y}")  # Output: 2

# 2. Exponent Notation Example
num = 1e3  # 1 * 10^3 = 1000
print(f"Exponent notation (1e3): {num}")  # Output: 1000.0

# 3. Compound Assignment Operators

# Example of Add and Assign (+=)
x = 10
x += 5  # Equivalent to x = x + 5
print(f"x += 5 gives: {x}")  # Output: 15

# Example of Subtract and Assign (-=)
x = 10
x -= 3  # Equivalent to x = x - 3
print(f"x -= 3 gives: {x}")  # Output: 7

# Example of Multiply and Assign (*=)
x = 4
x *= 3  # Equivalent to x = x * 3
print(f"x *= 3 gives: {x}")  # Output: 12

# Example of Divide and Assign (/=)
x = 20
x /= 4  # Equivalent to x = x / 4
print(f"x /= 4 gives: {x}")  # Output: 5.0

# Example of Modulus and Assign (%=)
x = 20
x %= 6  # Equivalent to x = x % 6
print(f"x %= 6 gives: {x}")  # Output: 2

# Example of Exponentiate and Assign (**=)
x = 2
x **= 3  # Equivalent to x = x ** 3
print(f"x **= 3 gives: {x}")  # Output: 8

# Example of Floor Divide and Assign (//=)
x = 20
x //= 6  # Equivalent to x = x // 6
print(f"x //= 6 gives: {x}")  # Output: 3
```
